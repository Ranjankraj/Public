```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url = "https://raw.githubusercontent.com/shala2020/shala2020.github.io/master/Lecture_Materials/Assignments/MachineLearning/L2/titanic.csv"

titanic = pd.read_csv(url)
titanic.head()
```

```
titanic.info()
```

```
titanic.isna().sum()
```

```
pd.set_option('display.max_rows', None)
print(titanic.Age)
```

```
titanic_1 = titanic.fillna(titanic.mean())
titanic_1.info()
```

```
titanic_1.drop(['Cabin'], axis=1, inplace=True)
titanic_1.info()
```

```
titanic_1.isna().sum()
```

```
titanic_1.dropna(inplace=True)
```

```
titanic_1.info()
```

```
titanic_2 = titanic.fillna(titanic.median())
print(titanic_2.info())
```

```
titanic_2.drop(['Cabin'], axis=1, inplace=True)
```

```
titanic_2.dropna(inplace=True)
print(titanic_2.info())
```

```
print(titanic_1.head())
print(titanic_1.Parch.value_counts())
```

```
titanic_1.drop(['Name', 'Ticket'], axis=1, inplace=True)
titanic_2.drop(['Name', 'Ticket'], axis=1, inplace=True)
print(titanic_1.head())
print(titanic_1.info())
```

```
titanic_1 = pd.get_dummies(titanic_1)
titanic_1.head()
```

```
titanic_2 = pd.get_dummies(titanic_2)
titanic_2.head()
```

```
X1 = titanic_1.drop(['PassengerId', 'Survived'], axis=1)
y1 = titanic_1['Survived']
X2 = titanic_2.drop(['PassengerId', 'Survived'], axis=1)
y2 = titanic_2['Survived']
```

```
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, stratify=y1, random_state=123)

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, stratify=y2, random_state=123)
```

```
# Import functions to compute accuracy and split data
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# Import models, including VotingClassifier meta-model
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import VotingClassifier
# Set seed for reproducibility
SEED = 123
```

```
lr = LogisticRegression(random_state=SEED)
knn = KNN()
dt = DecisionTreeClassifier(random_state=SEED)
# Define a list called classifier that contains the tuples (classifier_name, classifier)
classifiers = [('Logistic Regression', lr),
('K Nearest Neighbours', knn),
('Classification Tree', dt)]
```

```
import warnings
warnings.filterwarnings("ignore")


# Instantiate a VotingClassifier 'vc'
vc1 = VotingClassifier(estimators=classifiers)
# Fit 'vc' to the traing set and predict test set labels
vc1.fit(X1_train, y1_train)
y1_pred = vc1.predict(X1_test)
# Evaluate the test-set accuracy of 'vc'
print("Voting classifier 1: ", round(accuracy_score(y1_test, y1_pred),3))

# Instantiate a VotingClassifier 'vc'
vc2 = VotingClassifier(estimators=classifiers)
# Fit 'vc' to the traing set and predict test set labels
vc2.fit(X2_train, y2_train)
y2_pred = vc2.predict(X2_test)
# Evaluate the test-set accuracy of 'vc'
print("Voting classifier 2: ", round(accuracy_score(y2_test, y2_pred),3))
```

```
# Import models and utility functions
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# Set seed for reproducibility
SEED = 123
```

```
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=0.16, random_state=SEED)

# Instantiate a BaggingClassifier 'bc'
bc1 = BaggingClassifier(base_estimator=dt, n_estimators=300, n_jobs=-1) # n_jobs=-1 means that all the CPU cores are used in computation. 

# Fit 'bc' to the training set
bc1.fit(X1_train, y1_train)

# Predict test set labels
y1_pred = bc1.predict(X1_test)

# Evaluate and print test-set accuracy
accuracy1 = accuracy_score(y1_test, y1_pred)
print('Accuracy of Bagging Classifier: {:.3f}'.format(accuracy1))


# Instantiate a BaggingClassifier 'bc'
bc2 = BaggingClassifier(base_estimator=dt, n_estimators=300, n_jobs=-1) # n_jobs=-1 means that all the CPU cores are used in computation. 

# Fit 'bc' to the training set
bc2.fit(X2_train, y2_train)

# Predict test set labels
y2_pred = bc2.predict(X2_test)

# Evaluate and print test-set accuracy
accuracy2 = accuracy_score(y2_test, y2_pred)
print('Accuracy of Bagging Classifier: {:.3f}'.format(accuracy2))
```

```
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```
dt = DecisionTreeClassifier(max_depth=1, random_state=SEED)
# Instantiate an AdaBoost classifier 'adab_clf'
adb_clf1 = AdaBoostClassifier(base_estimator=dt, n_estimators=100)
# Fit 'adb_clf' to the training set
adb_clf1.fit(X1_train, y1_train)
# Predict the test set probabilities of positive class
y1_pred_proba = adb_clf1.predict_proba(X1_test)[:,1]
# Evaluate test-set roc_auc_score
adb_clf_roc_auc_score1 = roc_auc_score(y1_test, y1_pred_proba)
# Print adb_clf_roc_auc_score
print('ROC AUC score 1: {:.2f}'.format(adb_clf_roc_auc_score1))


adb_clf2 = AdaBoostClassifier(base_estimator=dt, n_estimators=100)
# Fit 'adb_clf' to the training set
adb_clf2.fit(X2_train, y2_train)
# Predict the test set probabilities of positive class
y2_pred_proba = adb_clf2.predict_proba(X2_test)[:,1]
# Evaluate test-set roc_auc_score
adb_clf_roc_auc_score2 = roc_auc_score(y2_test, y2_pred_proba)

print('ROC AUC score 2: {:.2f}'.format(adb_clf_roc_auc_score2))
```
